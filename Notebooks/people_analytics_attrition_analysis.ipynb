Total Employee Count Distribution

```python
%%sql

SELECT 
Department,
COUNT(*) AS employees
FROM employees
GROUP BY Department

UNION ALL

SELECT
'Grand Total' AS Department,  
COUNT(*) AS employees
FROM employees
ORDER BY CASE WHEN Department = 'Grand Total' THEN 2 ELSE 1 END, Department DESC;
```

Overall attrition rate

```python
%%sql
SELECT
COUNT(*) AS total_employees,
SUM(CASE WHEN Attrition = 'Yes' THEN 1 ELSE 0 END) AS total_attrition,
ROUND(SUM(CASE WHEN Attrition = 'Yes' THEN 1 ELSE 0 END) / COUNT(*) * 100, 2) AS attrition_rate
FROM employees;
```

Age Group Distribution

```python
%%sql
SELECT
    CASE 
        WHEN Age < 30 THEN 'Under 30'
        WHEN Age BETWEEN 30 AND 39 THEN '30-39'
        WHEN Age BETWEEN 40 AND 49 THEN '40-49'
        ELSE '50 and above'
    END AS age_group,
COUNT(*) AS employee_count
FROM employees
GROUP BY age_group
ORDER BY age_group;
```

Attrition Rate By Department. Find which groups have highest turnover

```python
%%sql
SELECT
    Department,
    COUNT(*) AS employee_count,
    SUM(CASE WHEN Attrition = 'Yes' THEN 1 ELSE 0 END) AS attrition_emp,
    ROUND(SUM(CASE WHEN Attrition = 'Yes' THEN 1 ELSE 0 END) / COUNT(*) * 100, 2) AS dept_attrition_pct
FROM employees
GROUP BY Department
ORDER BY dept_attrition_pct DESC;
```

Attrition Rate by Job Role

```python
%%sql
SELECT
    JobRole,
    COUNT(*) AS employee_count,
    SUM(CASE WHEN Attrition = 'Yes' THEN 1 ELSE 0 END) AS attrition_emp,
    ROUND(SUM(CASE WHEN Attrition = 'Yes' THEN 1 ELSE 0 END) / COUNT(*) * 100, 2) AS job_attrition_pct
FROM employees
GROUP BY JobRole
ORDER BY job_attrition_pct DESC;
```

Income vs Attrition (Salary Quartiles)

```python
%%sql
SELECT 
    q.income_quartile,
    MIN(q.MonthlyIncome) AS min_income,
    MAX(q.MonthlyIncome) AS max_income,
    ROUND(SUM(CASE WHEN q.Attrition = 'Yes' THEN 1 ELSE 0 END) / COUNT(*) * 100, 2) AS income_attrition_pct
FROM (
    SELECT 
        MonthlyIncome,
        Attrition,
        NTILE(4) OVER (ORDER BY MonthlyIncome) AS income_quartile
    FROM employees
) AS q
GROUP BY q.income_quartile;
```

Attrition by Years at Comoany

```python
%%sql
SELECT
CASE 
    WHEN YearsAtCompany < 1 THEN 'Less than 1 year'
    WHEN YearsAtCompany BETWEEN 1 AND 3 THEN '1-3 years'
    WHEN YearsAtCompany BETWEEN 4 AND 6 THEN '4-6 years'
    WHEN YearsAtCompany BETWEEN 7 AND 10 THEN '7-10 years'
    ELSE '10+ years'
END AS tenure_group,
COUNT(*) AS employee_count,
ROUND(SUM(CASE WHEN Attrition = 'Yes' THEN 1 ELSE 0 END) / COUNT(*) * 100, 2) AS tenure_attrition_pct
FROM employees
GROUP BY tenure_group
ORDER BY tenure_attrition_pct DESC;
```

High Risk Employees. Identify high rish employees for retention

Attrition Risk Score

I build a simple attrition risk score based on known risk factors:

Overtime = 30 points
Low tenure = 25 points
Low satisfaction = 20 points
Low salary = 15 points
MaritalStatus Single = 10

```python
%%sql
SELECT
EmployeeNumber,
JobRole,
Department,
(CASE WHEN OverTime = 'Yes' THEN 30 ELSE 0 END +
CASE WHEN YearsAtCompany < 3 THEN 25 ELSE 0 END +
CASE WHEN JobSatisfaction <= 2 THEN 20 ELSE 0 END +
CASE WHEN MonthlyIncome < 5000 THEN 15 ELSE 0 END +
CAsE WHEN MaritalStatus = 'Single' THEN 10 ELSE 0 END) AS attrition_risk_score
FROM employees
where Attrition = 'No'
ORDER BY attrition_risk_score DESC;
```
